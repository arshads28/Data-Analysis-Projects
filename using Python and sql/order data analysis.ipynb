```python
df= pd.read_csv(r'C:\Users\arsh\Downloads\orders.csv' ,na_values =['Not Available', 'unknown' ] )
df['Ship Mode'].unique()
```

```python

# columns values are not good match
df.columns=df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.head()
```

```python
# find tha discount
df["discount"]=df['list_price']*df['discount_percent']*0.01
```

```python
#find the sale proce
df['sale_price'] = df["list_price"]-df['discount']
```

```python
#find th profit
df['profit'] = df['sale_price']-df['cost_price']
```
